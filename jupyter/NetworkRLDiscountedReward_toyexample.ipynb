Construct a network graph and compute the d-hop neighborhood of each node. Use addEdge(i, j) to add an undirected edge between node i and j. After finishing the process of adding edges, call finishAddingEdges, so the object will construct d-hop adjacency matrices for $d \leq k$. Call findNeighbors(i, d) to get the d-hop neighborhood of node i.

```python
class GlobalNetwork:
    def __init__(self, nodeNum, k):
        self.nodeNum = nodeNum #the total number of nodes in this network
        self.adjacencyMatrix = np.eye(self.nodeNum, dtype = int) #initialize the adjacency matrix of the global network
        self.k = k #the number of hops used in learning
        self.adjacencyMatrixPower = [np.eye(self.nodeNum, dtype = int)] #cache the powers of the adjacency matrix
        self.neighborDict = {} #use a hashmap to store the ((node, dist), neighbors) pairs which we have computed
        self.addingEdgesFinished = False #have we finished adding edges?
    
    #add an undirected edge between node i and j
    def addEdge(self, i, j):
        self.adjacencyMatrix[i, j] = 1
        self.adjacencyMatrix[j, i] = 1
    
    #finish adding edges, so we can construct the k-hop neighborhood after adding edges
    def finishAddingEdges(self):
        temp = np.eye(self.nodeNum, dtype = int)
        #the d-hop adjacency matrix is stored in self.adjacencyMatrixPower[d]
        for _ in range(self.k):
            temp = np.matmul(temp, self.adjacencyMatrix)
            self.adjacencyMatrixPower.append(temp)
        self.addingEdgesFinished = True
    
    #query the d-hop neighborhood of node i, return a list of node indices.
    def findNeighbors(self, i, d):
        if not self.addingEdgesFinished:
            print("Please finish adding edges before call findNeighbors!")
            return -1
        if (i, d) in self.neighborDict: #if we have computed the answer before, return it
            return self.neighborDict[(i, d)]
        neighbors = []
        for j in range(self.nodeNum):
            if self.adjacencyMatrixPower[d][i, j] > 0: #this element > 0 implies that dist(i, j) <= d
                neighbors.append(j)
        self.neighborDict[(i, d)] = neighbors #cache the answer so we can reuse later
        return neighbors
```

```python
def constructLinearNetwork(nodeNum, k):
    #compute the number of access points
    
    network = GlobalNetwork(nodeNum = nodeNum, k = k)
    
    for i in range(nodeNum):
        if i >= 1:
            network.addEdge(i, i-1)
        if i < nodeNum-1:
            network.addEdge(i, i+1)
    network.finishAddingEdges()    
    return network
```

Class Node specifies what should be implemented by a (user) node in the global network. It maintains the history record of its local trajectory as well as its k-hop neighbors' (state, action) trajectory in the learning process.

The offspring of this class should implement:
1. initializeState: pick the initial state of this (user) node;
2. updateState: based on neighbors' states and actions at the last time step, pick the current local state;
3. updateAction: based on the local state (or neighbors' states), pick the current action;
4. updateReward: based the neighbors' current states and actions, decide the current local reward;
5. updateQ: based on $(s_{N_i^k}(t-1), a_{N_i^k}(t-1), r_i(t-1), s_{N_i^k}(t), a_{N_i^k}(t))$, update the Q function;
6. updateParams: using the recorded trajectory and the Q dictionary to update the policy parameters.

```python
class Node:
    def __init__(self, index):
        self.index = index
        self.state = [] #The list of local state at different time steps
        self.action = [] #The list of local actions at different time steps
        self.reward = [] #The list of local actions at different time steps
        self.paramsDict = {} #use a hash map to query the parameters given a state (or neighbors' states)
        self.QDict = {} #use a hash map to query to the Q value given a (state, action) pair
        self.kHop = [] #The list to record the (state, action) pairs of k-hop neighbors
        self.currentTimeStep = 0
    #get the local state at timeStep
    #get the local Q at timeStep
    def getQ(self, kHopStateAction):
        #if the Q value of kHopStateAction hasn't been queried before, return 0.0 (initial value)
        return self.QDict.get(kHopStateAction, 0.0)
    
    #initialize the local state
    def initializeState(self):
        pass
    #update the local state, it may depends on the states of other nodes at the last time step.
    #Remember to increase self.currentTimeStep by 1
    def updateState(self):
        pass
    #update the local action
    def updateAction(self):
        pass
    #update the local reward
    def updateReward(self):
        pass
    #update the local Q value
    def updateQ(self):
        pass
    #update the local parameter
    def updateParams(self):
        pass
    #clear the record. Called when a new inner loop starts. 
    def restart(self, clearPolicy = False, clearQ = False):
        self.state.clear()
        self.action.clear()
        self.reward.clear()
        
        if(clearPolicy == True):
            self.paramsDict.clear()
        if(clearQ == True):
            self.QDict.clear()
        self.kHop = []
        self.currentTimeStep = 0
```

```python
class ToyNode(Node):
    def __init__(self, stateNum, actionNum, nodeNum, index, neighborStateShape, minProb, gamma):
        super(ToyNode, self).__init__(index)
        #we use packetQueue to represent the current local state, which is (e_1, e_2, ..., e_d)
        self.actionNum = stateNum #the number of possible actions
        self.stateNum = actionNum # number of possible states
        self.nodeNum = nodeNum
        self.gamma = gamma
        # local reward        
        self.rewardFunc = np.random.rand(self.stateNum,self.actionNum)
        
        # the following is for testing !!!
        #self.rewardFunc = np.random.rand(self.stateNum,self.actionNum)*(index==0)*10
        if index == 0:
            self.rewardFunc[:,: ]= 0
            self.rewardFunc[1,: ]= 1 # one state has high reward
        else:
            self.rewardFunc[:,: ]= 0.0
        # ------------ testing end ---------
        
        
        self.shapeList = [self.stateNum] + neighborStateShape + [self.actionNum] 
        # local transition 
        tmpLocalP = np.random.rand(*tuple(self.shapeList))
        normalizationP = np.expand_dims(np.sum(tmpLocalP,axis = 0),axis = 0)
        self.localP = (1 - self.stateNum* minProb )*tmpLocalP/ np.repeat(normalizationP, self.stateNum, axis = 0) + minProb
        
        # the following is for testing!!!
        if index == 0:
            self.localP[0,:,1,:] =  0.0#self.localP[0,:,1,:]/20 # if right neighbor is 1, then has high prob. of being 1
            self.localP[1,:,1,:] = 1.0
            
            self.localP[0,:,0,:] = 1.0
            self.localP[1,:,0,:] = 0.0
        elif index >= 1 and index< nodeNum -1: # not the right most node
#             self.localP[0,:,:,:,0] = 0.99
#             self.localP[1,:,:,:,0] = 0.01

#             self.localP[0,:,:,:,1] = 0.01
#             self.localP[1,:,:,:,1] = 0.99
            
            self.localP[1,:,:,1,0] = 0.0 # action 0, low prob. of state 1
            self.localP[0,:,:,1,0] = 1.0

            self.localP[1,:,:,1,1] = 1.0 # action 1, high prob. of state 0
            self.localP[0,:,:,1,1] = 0.0

            self.localP[1,:,:,0,0] = 0.0 # action 0, low prob. of state 1
            self.localP[0,:,:,0,0] = 1.0

            self.localP[1,:,:,0,1] = 0.8 # action 1, low prob. of state 0
            self.localP[0,:,:,0,1] = 0.2

        else:# last node
            self.localP[0,:,:,0] = 1.0
            self.localP[1,:,:,0] = 0.0

            self.localP[0,:,:,1] = 0.0
            self.localP[1,:,:,1] = 1.0
           
        self.defaultPolicy = np.array([-0.2,0.2])
        #tmpLocalP = np.exp(np.random.rand(*tuple(self.shapeList)) - 0.5)*4
        #normalizationP = np.expand_dims(np.sum(tmpLocalP,axis = 0),axis = 0)
        #self.localP = (1 - self.stateNum* minProb )*tmpLocalP/ np.repeat(normalizationP, self.stateNum, axis = 0) + minProb
            
        
        # ------------ testing end ---------
        
        
    #initialize the local state (called at the beginning of the training process)
    def initializeState(self):
        self.state.append(1) #append this initial state to state record
        
    #At each time step t, call updateState, updateAction, updateReward, updateQ in this order
    def updateState(self,neighborStateList):
        self.currentTimeStep += 1
        lastAction = self.action[-1]
        probVec = []
        
        for nextState in range(self.stateNum):
            query = [nextState] + neighborStateList + [lastAction]
            probVec.append(self.localP[ tuple(query) ])
            
        newState = np.random.choice(a = self.stateNum, p = probVec)
        self.state.append(newState) #append this state to state record
                    
        
    def updateAction(self):
        # get the current state
        currentState = self.state[-1]
        # fetch the params based on the current state. If haven't updated before, return all zeros
        params = self.paramsDict.get(currentState, self.defaultPolicy)
        # compute the probability vector
        probVec = special.softmax(params)
        # randomly select an action based on probVec
        currentAction = np.random.choice(a = self.actionNum, p = probVec)
        self.action.append(currentAction)
    
    #oneHopNeighbors is a list of accessNodes
    def updateReward(self):
        currentAction = self.action[-1]
        currentState = self.state[-1]
        
        self.reward.append(self.rewardFunc[currentState,currentAction])

    
    #kHopNeighbors is a list of accessNodes, alpha is learning rate
    def updateQ(self, kHopNeighbors, alpha):
        lastStateAction = []
        currentStateAction = []
        #construct a list of the state-action pairs of k-hop neighbors
        dummyFlag = []
        for neighbor in kHopNeighbors:
            neighborLastState = neighbor.state[-2]
            neighborCurrentState = neighbor.state[-1]
            neighborLastAction = neighbor.action[-2]
            neighborCurrentAction = neighbor.action[-1]
            lastStateAction.append((neighborLastState, neighborLastAction))
            currentStateAction.append((neighborCurrentState, neighborCurrentAction))
        lastStateAction = tuple(lastStateAction)
        currentStateAction = tuple(currentStateAction)
        #fetch the Q value based on neighbors' states and actions
        lastQTerm1 = self.QDict.get(lastStateAction, 0.0)
        lastQTerm2 = self.QDict.get(currentStateAction, 0.0)
        #compute the temporal difference
        
        temporalDiff = self.reward[-2] + self.gamma* lastQTerm2 - lastQTerm1
        #print('lastStateAction',lastStateAction)
        self.QDict[lastStateAction] = lastQTerm1 + alpha * temporalDiff
            #print(self.QDict[lastStateAction])
        # if this time step 1, we should also put lastStateAction into history record
        
        if len(self.kHop) == 0:
            self.kHop.append(lastStateAction)
        #put currentStateAction into history record
        self.kHop.append(currentStateAction)
    
    #eta is the learning rate
    def updateParams(self, kHopNeighbors):
        #for t = 0, 1, ..., T, compute the term in g_{i, t}(m) before \nabla
        mutiplier1 = np.zeros(self.currentTimeStep + 1)
        for neighbor in kHopNeighbors:
            for t in range(self.currentTimeStep + 1):
                neighborKHop = neighbor.kHop[t]
                neighborQ = neighbor.getQ(neighborKHop)
                mutiplier1[t] += neighborQ
        for t in range(self.currentTimeStep + 1):
            mutiplier1[t] *= pow(self.gamma, t)
            mutiplier1[t] /= nodeNum
        
        total_grad = np.zeros((self.stateNum,self.actionNum))
        for t in range(self.currentTimeStep + 1):
            currentState = self.state[t]
            currentAction = self.action[t]
            params = self.paramsDict.get(currentState, self.defaultPolicy)
            probVec = special.softmax(params)
            grad = -probVec
            grad[currentAction] += 1.0
            total_grad[currentState,:] = total_grad[currentState,:] + mutiplier1[t] * grad
            #self.paramsDict[currentState] = params + eta * mutiplier1[t] * grad
                
        return total_grad
        
```

```python
class NetworkRLToy:
    def __init__(self,nodeNum,kMax,S = 2,A=2,minProb =0.0,gamma = 0.7):
        self.nodeNum = nodeNum
        self.S = S
        self.A = A
        self.network = constructLinearNetwork(nodeNum,kMax)
        self.gamma = gamma
        # construct neighbor list
        self.neighborList = []
        self.shapeList = []
        self.nodeList = []
        for i in range(nodeNum):
            self.neighborList.append(self.network.findNeighbors(i,1))
            self.shapeList.append( [self.S]*len(self.neighborList[i]))
            self.nodeList.append( ToyNode(stateNum= self.S, actionNum = self.A, nodeNum = nodeNum, index = i, neighborStateShape = self.shapeList[i],minProb = minProb ,gamma=self.gamma))
    def getNeighborState(self,i):
        neighborState = []
        for j in self.neighborList[i]:
            neighborState.append(self.nodeList[j].state[-1]) # append neighbor's state
        return neighborState
    
        
    def train(self, k = 1, M = 10000, T = 20, evalInterval = 500, restartIntervalQ = 50, restartIntervalPolicy = 50, evalM = 100, actorQuietPeriod = 0, batchSize = 10):
        policyRewardList = []
        policyRewardSmooth = []
        
        batchGrad = []
        
        for i in range(self.nodeNum):
            batchGrad.append(np.zeros((self.S,self.A)))
        
        for m in trange(M):
            # restart policy 
            discountedReward = 0.0
            tmpReward = 0.0
            for i in range(self.nodeNum):
                if(m==0):
                    self.nodeList[i].restart(clearPolicy=True,clearQ = True)
                else:
                    self.nodeList[i].restart(clearPolicy=False,clearQ = False)
                self.nodeList[i].initializeState()
                self.nodeList[i].updateAction()
                self.nodeList[i].updateReward()
                tmpReward += self.nodeList[i].reward[-1]
        
            discountedReward += tmpReward/self.nodeNum
            
            for t in range(T):
                tmpReward = 0.0
                
                neighborStateList = []
                for i in range(self.nodeNum): 
                    neighborStateList.append(self.getNeighborState(i))
                
                for i in range(self.nodeNum): #update state-action
                    self.nodeList[i].updateState(neighborStateList[i])
                    self.nodeList[i].updateAction()
                    self.nodeList[i].updateReward()
                    tmpReward += self.nodeList[i].reward[-1] # add latest reward
                discountedReward += (self.gamma**(t+1))*tmpReward/self.nodeNum
                
                for i in range(self.nodeNum):
                    neighborList = []
                    for j in self.network.findNeighbors(i, k):
                        neighborList.append(self.nodeList[j])
                    self.nodeList[i].updateQ(neighborList, 1/pow(( ((m-1)^T+t)%restartIntervalQ)+1,.5)  )

            

            # perform the grad update
            
            if m<=actorQuietPeriod:
                actorStepSize = 0.0
            else:
                actorStepSize = 0.1* 1/pow(( ((m - actorQuietPeriod)//batchSize)%restartIntervalPolicy)+1,.5)
            
            
            for i in range(self.nodeNum):
                neighborList = []
                for j in self.network.findNeighbors(i, k):
                    neighborList.append(self.nodeList[j])
                batchGrad[i] =batchGrad[i] + self.nodeList[i].updateParams(neighborList ) # returns gradient
            
            # after batch size reached, average over batched gradient and do gradient descent
            if m%batchSize == batchSize - 1:
                for i in range(self.nodeNum):
                    batchGrad[i] = batchGrad[i]/batchSize # average
                    for state in range(self.S): # gradient descent
                        self.nodeList[i].paramsDict[state] = self.nodeList[i].paramsDict.get(state,  self.nodeList[i].defaultPolicy)+ actorStepSize*batchGrad[i][state,:]
                    #print('batch gradient of node ', str(i), ' is:',batchGrad[i])
                    batchGrad[i] = np.zeros((self.S,self.A))# clear gradient
            
            
           
            # print(m%restartIntervalPolicy)
            if m > M*0.9: # for the last 10% of running, no restarting
                restartIntervalQ = max(int(M*0.5),restartIntervalQ)*T
                restartIntervalPolicy = max(int(M*0.5),restartIntervalPolicy)
            
            # perform a policy evaluation
            policyRewardList.append(tmpReward/self.nodeNum) # update reward 

            if m%evalInterval == evalInterval - 1:
                policyRewardSmooth.append(self.policyEval(evalM = evalM, T=T) )
                for i in range(self.nodeNum):
                    pass
                    #print('node ',str(i),' Q value: ',self.nodeList[i].QDict)
                    #print('node ',str(i),' policy: ',self.nodeList[i].paramsDict)
                #for i in range(self.nodeNum):
                #    print('policy of node #'+str(i)+' is: ', self.nodeList[i].paramsDict)
            
        return policyRewardSmooth
    
    def policyEval(self, evalM, T):
        aveReward = 0.0
        for m in range(evalM):
            # restart policy 
            discountedReward = 0.0
            tmpReward = 0.0
            for i in range(self.nodeNum):
                self.nodeList[i].restart(clearPolicy=False,clearQ = False)
                self.nodeList[i].initializeState()
                self.nodeList[i].updateAction()
                self.nodeList[i].updateReward()
                tmpReward += self.nodeList[i].reward[-1]
        
            discountedReward += tmpReward/self.nodeNum
            
            for t in range(T):
                tmpReward = 0.0
                neighborStateList = []
                for i in range(self.nodeNum): 
                    neighborStateList.append(self.getNeighborState(i))
                
                for i in range(self.nodeNum): #update state-action
                    self.nodeList[i].updateState(neighborStateList[i])
                    self.nodeList[i].updateAction()
                    self.nodeList[i].updateReward()
                    tmpReward += self.nodeList[i].reward[-1] # add latest reward
                discountedReward += (self.gamma**(t+1))*tmpReward/self.nodeNum
                
            aveReward += discountedReward
        return aveReward/evalM
        
        

```

```python

#3 node linear network
simuCase = 'shortline'
nodeNum = 8 #number of nodes in the network
kMax = 7 #the size of neighborhood we use in localized learning
gamma = 0.7
np.random.seed(0)
model = NetworkRLToy(nodeNum = nodeNum,kMax = kMax,S=2,A=2, minProb = .0, gamma = gamma)

M = 10000
T = 10

evalInterval = 200 #evaluate the policy every evalInterval rounds (outer loop)
evalM = 500 # number of episodes when doing evaluation
batchSize = 1 # batch size for doing gradient descent
restartIntervalQ = 50
restartIntervalPolicy = 50

np.random.seed()

```

```python
policyRewardSmoothList = []
finalRewardList = []

for k in range(kMax+1):
    actorQuietPeriod = 100
    policyRewardSmooth = model.train(k = k, M = M, T=T, evalInterval = evalInterval,restartIntervalQ = restartIntervalQ\
                                 , restartIntervalPolicy = restartIntervalPolicy, evalM = evalM, actorQuietPeriod = actorQuietPeriod,\
                                    batchSize = batchSize)
    policyRewardSmoothList.append(policyRewardSmooth)
    finalRewardList.append(policyRewardSmooth[-1])
    
    for i in range(nodeNum):
        #print('k = ', str(k), ', i = ', str(i), ' QDict = ', model.nodeList[i].QDict)
        print('k = ', str(k), ', i = ', str(i), ' policy = ', model.nodeList[i].paramsDict)
```

```python
plt.rc('font', size=13)  

finalRewardList = []
legendStr = []
optReward = (1 - gamma**(T+1))/(float(nodeNum)*(1-gamma))
for kInd in range(kMax+1):
    plt.plot(np.linspace(1,M,len(policyRewardSmooth)), policyRewardSmoothList[kInd])
    legendStr.append('$\kappa = '+str(kInd)+'$')
    
    finalRewardList.append(np.mean(np.array(policyRewardSmoothList[kInd][-10:-1])))
plt.plot(np.linspace(1,M,len(policyRewardSmooth)),np.ones(len(policyRewardSmooth),dtype = float)*optReward,color = 'black',linestyle = ':'  )
legendStr.append('Optimal')
plt.legend(legendStr)

plt.xlim(0,M)
plt.ylabel('Total Discounted Reward')
plt.xlabel('$m$ (training iterations)')
plt.savefig('Aug21Figs/synthetic_training_line8.png',bbox_inches='tight')
plt.show()

finalRewardList = np.array(finalRewardList)
plt.plot(range(kMax+1), finalRewardList)
plt.xlabel('$\kappa$')
plt.ylabel('Final Reward Achieved')

plt.show()


plt.plot(range(kMax+1), optReward - finalRewardList )
plt.xlabel('$k$')
plt.ylabel('Optimality Gap')
#plt.yticks(np.array([1e-3, 1e-1]))
plt.yscale('log')
plt.savefig('Aug21Figs/synthetic_exponential_decay_line8.png',bbox_inches='tight')
plt.show()
```
